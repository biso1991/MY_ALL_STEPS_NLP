```python
f = open('6.1.5 Tweets.json',encoding = 'utf-8')
data = json.load(f)  
```

```python
AllKeys = []
for i in data : 
    AllKeys.extend(i.keys())
set(AllKeys)    
```

```python
Data = pd.DataFrame(columns = set(AllKeys))
Data.head()
```

```python
for j,i in enumerate(data) :
    Data.loc[j] = 0
    for a,b in i.items() : 
        Data.loc[j][a] = str(b)
```
