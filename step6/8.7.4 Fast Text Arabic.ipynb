```python
import io
```

```python
def load_vectors(fname,length = 10000):
    fin = io.open(fname, 'r', encoding='utf-8', newline='\n', errors='ignore')
    n, d = map(int, fin.readline().split())
    data = {}
    i=0
    for line in fin:
        i+=1
        tokens = line.rstrip().split()
        data[tokens[0]] = tokens[1:]
        if i == length : 
            break
    return data
```

```python
ArabicEmbedding = load_vectors('../input/fasttext-pretrained-arabic-word-vectors/cc.ar.300.vec')
```

```python
def SimilarWords(Word) :
    if Word in ArabicEmbedding.keys() : 
        WordsDifference = {}
        for EachWord in ArabicEmbedding.keys() : 
            WordsDifference[EachWord] = sum([abs(float(i)-float(j)) for i,j in zip(ArabicEmbedding[Word],ArabicEmbedding[EachWord])])
        return list({k: v for k, v in sorted(WordsDifference.items(), key=lambda item: item[1])}.items())[:10]
    else : 
        return 'Word not exists'
```
